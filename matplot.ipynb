```python
data=pd.DataFrame({"salary":[2500,30000,37000,28000,39000,48000,5500,52000,35000,26000,27000,31000,35000]})
```

```python
# line plot
plt.plot(data, color="red",marker="o", linestyle="--",linewidth="2")
plt.grid(True)
plt.show()


```

```python
# histogram
plt.hist(data,bins=5,color="green")
plt.show()
```

```python
# box plot
plt.boxplot(data)
plt.show()
```

```python
# univariate categorical value
data["dept"]=['HR','Finance','IT','IT','HR','HR','HR','Finance','IT','IT','HR','Finance','Finance']
data.head()
```

```python
# inorder to visualize categorical(string) column we use pie chart
count=data["dept"].value_counts()
count
```

```python
plt.pie(count,labels=count.index ,autopct="%1.2f",explode=[0,0.2,0.2])
plt.axis("equal")
plt.show()
```

```python
# next way is count plot
plt.bar(count.index,count,color=["red","black","green"])
plt.grid(True)
plt.show()
```

```python
# if we have 2 cloumn at a time
# -Bivariate analysis
# numerical-numerical
data["age"]=[22,20,46,29,32,45,20,46,29,40,25,30,56]
data.head()
```

```python
# scatter plot
plt.scatter(data["salary"],data["age"],color="orange")
plt.show()
```

```python
# line plot--(not easy to visualize becoz age is not sorted)
plt.plot(data["salary"],data["age"], marker="*")
plt.grid(True)
plt.show()
```

```python
sort_age=data.sort_values("age")
plt.plot(data["salary"],sort_age["age"], marker="*")
plt.grid(True)
plt.show()
```

```python
# bar chart
plt.bar(data["age"],data["salary"],color="green")
plt.show()
```

```python
# to visualize : numeriacl +categorical column
hr_sal=data[data["dept"]=="HR"]["salary"]
hr_sal
```

```python
it_sal=data[data["dept"]=="IT"]["salary"]
it_sal
```

```python
fin_sal=data[data["dept"]=="Finance"]["salary"]
fin_sal
```

```python
plt.boxplot([hr_sal,it_sal,fin_sal],tick_labels=["HR","IT","finance"])
plt.show()
```

```python
# pie chart
salary_by_dept=data.groupby("dept")["salary"].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept ,labels=salary_by_dept.index,autopct="%1.2f")
plt.show()
```

```python
# bar plot
plt.bar(salary_by_dept.index,salary_by_dept)
plt.grid(True)
plt.show()
```

```python
# multi varaite analysis
# for 3 numerical
data.head()
```

```python
data["exp"]=[5,10,2,6,9,12,3,5,1,8,7,2,4]
data.head()
```

```python
# bubble plot-larger the bubble larger the experience
plt.scatter(data["age"],data["salary"],s=data["exp"]*15,color="orange",edgecolors="black")
plt.title("Age vs Salary vs Experience")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

```python
# 2 num + 1 categorical
plt.scatter(data["age"],data["salary"],c=data["dept"].map({"HR":"yellow","IT":"blue","Finance":"orange"}))
plt.title("Age vs Salary vs dept")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.legend()
plt.show()
```

```python
color={"HR":"yellow","IT":"blue","Finance":"orange"}
for dept,color in color.items():
    data_dept=data[data["dept"]==dept]
    plt.scatter(data_dept["age"],data_dept["salary"],c=color,label=dept)
    plt.legend()
```
