```python
sales_data=np.array([
    [1,150,180,2200,2500],  #paradise biryani
    [2,1200,1400,1600,1900], #beijing bites
    [3,2000,2300,2600,3000], #pizza hub
    [4,1800,2100,2400,2700], #burger point
    [5,100,18500,20500,2300] # chai point
])
```

```python
print(sales_data.shape)
# print(sales_data[0:3])
# total sales per year
print(np.sum(sales_data[:,1:],axis=1))
yearly=np.sum(sales_data[:,1:],axis=1)
print(np.sum(yearly))

# min sales per resturant
print("min sales per resturant")
print(np.min(sales_data[:,1:],axis=1))

# max sales per year
print("max sales per year")
print(np.max(sales_data[:,1:],axis=0))

# avg sales per resturant
print("avg sales per resturant")
print(np.mean(sales_data[:,1:],axis=1))

# cumulative sales per resturant
print("cumulative sales per resturant")
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)

print("cumulative graph")
plt.Figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales across all resturant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
